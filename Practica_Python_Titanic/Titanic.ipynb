## 2. Cargamos los datasets

#### Leemos los ficheros con el paquete csv 

```python
import pandas as pd
import csv
```

```python
test_df = pd.DataFrame.from_csv('datasets/titanic/test.csv')
train_df = pd.DataFrame.from_csv('datasets/titanic/train.csv')
gender_submission_df = pd.DataFrame.from_csv('datasets/titanic/gender_submission.csv')
```

```python
#Como el Dataset viene dividido vamos a juntarlo para hacer las operaciones de limpieza
#Añadimos la columna de Surviver a test_df
test_df['Survived']= gender_submission_df
```

```python
#Unimos los dos dataframe
frames = [test_df,train_df]
result = pd.concat(frames)
#Lo ordenamos por el index ('PassengerId')
result.sort_index(inplace=True)
#Ponemos como index el nombre del pasajero
id_serie = result.index.tolist()
names_serie = result['Name']
result.drop('Name', axis=1, inplace=True)
result.set_index([names_serie],inplace=True)
```

```python
result['PassengerId'] = id_serie
```

```python
result
```

```python
result.drop(['Cabin','Ticket'], axis=1, inplace=True)
```

## 5. Clasificadores

### 5.1 DecisionTreeClasifier

#### Cogemos el 80% para train y el 20% restante para test

```python
from sklearn.cross_validation import train_test_split
X = result[['Age','Embarked','Fare','Parch','Pclass','SibSp','female','male']] #Quitamos el ID porque no nos dice nada
y = result['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn import tree

my_tree = tree.DecisionTreeClassifier()
my_tree.fit(X_train, y_train)
```

```python
X_test['SURV PRED'] = my_tree.predict(X_test)
X_test['SURV REAL']= y_test
X_test['PRED'] = X_test['SURV PRED']==X_test['SURV REAL']
```

```python
X_test.groupby(['SURV REAL','SURV PRED']).count()
```

```python
accuracy_tree = 100 * float((X_test['PRED']==True).sum()) / X_test['PRED'].count()
print 'Porcentaje de aciertos --> ' +("%.2f" % accuracy_tree) + '%'
```
